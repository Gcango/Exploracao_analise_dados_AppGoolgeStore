```python
#1) - Qual é o total de aplicativos com títulos que contêm astrologia?
```

```python
dados.head(3)
```

```python
app_astr = dados[dados['App'].str.contains('Astrology', case=False)]
print(f"No total são {len(app_astr)} App com titulo de Astrology")
```

```python
#2) - Qual é a classificação média dos aplicativos?
```

```python
media = dados['Rating'].mean()
print(f"A média das classificações dos Aplicativos e de  {media:.4f} ")
```

```python
#3) - Qual é o numero total de categorias dos aplicativos?
```

```python
num_total = dados['Category'].nunique()
print(f"O total de categoria dos aplicativos e {num_total} ")
```

```python
#4) - Qual categoria obtem a classificação média mais alta?
```

```python
categ = dados.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print(f"São abaixo descritas as categorias com maior classificação: \n{categ} ")
```

```python
#5) - Qual é o número total de aplicativos com classificação de 5 estrelas?
```

```python
cinco_star = dados['Rating'].value_counts().sort_index(ascending=False)
print(f"O total de aplicativos com 5 estrelas foram {cinco_star.iloc[1]} ")
```

```python
#6) - Qual é o valor médio das avaliações?
```

```python
#dados['Reviews'].astype('float') este comando descobriu um alfanumerico nesta coluna
```

```python
dados[dados['Reviews'] == '3.0M']
```

```python
dados['Reviews'] = dados['Reviews'].replace('3.0M', 3.0)
```

```python
dados['Reviews'] = dados['Reviews'].astype('float')
```

```python
dados['Reviews'].dtype
```

```python
media_avalia = dados['Reviews'].mean()
print(f"A média das avaliações dos aplicativos foi {media_avalia:.4f} ")
```

```python
#7) - Qual é o número total de aplicativos gratuitos e pagos?
```

```python
app_type = dados['Type'].value_counts()
print(f"O numero Total dos aplicativos Gratis é de {app_type[0]} e o Pago é de {app_type[1]} ")
```

```python
#8) - Quais aplicativos possuem avaliaçao máxima?
```

```python
max_av = dados[dados['Reviews'].max() == dados['Reviews']]['App']
print(f"O aplicativo com avaliação máxima é: \n{max_av} ")
```

```python
#9) - Mostre os 5 aplicativos com melhores avaliações?
```

```python
print(f"Os 5 aplicativo com as melhores avaliações são: ")
dados[['Reviews','App']].sort_values(by = 'Reviews', ascending=False).head()
```

```python
#10) - Qual é a média das classificações dos aplicativos gratis e pagos?
```

```python
media_cl = dados.groupby('Type')['Rating'].mean()
print(f"A média das classificações dos aplicativos Gratis foi {media_cl[1]:.4f} e dos aplicativos Pagos {media_cl[2]:.4f} ")
```
